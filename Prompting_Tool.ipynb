```python
# @title
from transformers import CLIPTokenizer
import nltk
from nltk.corpus import stopwords
from transformers import CLIPTokenizer

# Download stopwords if not already present
nltk.download('stopwords')

# Initialise CLIP tokeniser
tokenizer = CLIPTokenizer.from_pretrained("openai/clip-vit-base-patch32")
stop_words = set(stopwords.words('english'))
```

## **Check Token Number**

```python
input = "This \"interpolatopia\" is a state-of-the-art work." #@param{type:'string'}
tokens = tokenizer.tokenize(input)
print(tokens)
print("Number of tokens:", len(tokens))
```

## **Prompt Optimiser**

```python
# @title
def optimise_prompt(input_text):
    """
    Optimises the given prompt by removing unnecessary words (e.g., stopwords)
    while keeping the original meaning as much as possible.

    Parameters:
    input_text (str): The input text to be optimised.

    Returns:
    str: The optimised text with minimal words.
    """

    # Step 1: Tokenise the input text into words
    words = input_text.split()

    # Step 2: Remove stopwords (like "the", "and", etc.)
    optimised_words = [word for word in words if word.lower() not in stop_words]

    # Step 3: Join the words back into a string
    optimised_text = ' '.join(optimised_words)

    # Return the optimised text
    return optimised_text


# Optimise the prompt
optimised_text = optimise_prompt(input)

# Print the original text
print("Original Text:", input)

# Print the label in default color and the optimised text in red
print("Optimised Text: \033[93m" + optimised_text + "\033[0m")

# Tokenise the original and optimised text
tokens_original = tokenizer.tokenize(input)
tokens_optimised = tokenizer.tokenize(optimised_text)

# Debugging: Check the number of tokens and the tokens themselves
print(f"Original Tokens ({len(tokens_original)}):", tokens_original)
print(f"Optimised Tokens ({len(tokens_optimised)}):", tokens_optimised)

# Check if optimised version has fewer tokens
if len(tokens_optimised) < len(tokens_original):
    print("Optimisation successful: Fewer tokens in the optimised prompt.")
else:
    print("Optimisation resulted in more tokens or the same number of tokens.")
```
